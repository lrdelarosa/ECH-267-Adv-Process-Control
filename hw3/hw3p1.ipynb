```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
# Lyapunov-based controller (let c = 2)
x = np.linspace(0, 10, 100)
u1 = 3*x #abs value
# Sontag's controller
u2 = x + x*np.sqrt(1+x**4) #abs value
```

```python
# Plot u vs. x
plt.plot(x, u1, 'b', label = 'Lyapunov-based controller')
plt.plot(x, u2, 'r', label = 'Sontag controller')
plt.legend(loc='best')
plt.xlabel('|x|')
plt.ylabel('|u|')
plt.title('Lyapunov-based controller vs. Sontag controller')
plt.show
```

****Problem 1 Part d****

Consider three different initial conditions: x0 = 1, x0 = 4, and x0 = 7 

```python
# For the Lyapunov-based controller with c = 2
def f(x, u):
    dxdt = -x**3
    return dxdt

# Solve the ODE with the initial conditions 
x0_1 = 1
x0_2 = 4
x0_3 = 7
t = np.linspace(0, 10, 100)
x1 = odeint(f, x0_1, t)
x2 = odeint(f, x0_2, t)
x3 = odeint(f, x0_3, t)

# Plot x vs. t
plt.plot(t, x1, 'b', label = 'x(0) = 1')
plt.plot(t, x2, 'r', label = 'x(0) = 4')
plt.plot(t, x3, 'g', label = 'x(0) = 7')
plt.legend(loc='best')
plt.xlabel('t')
plt.ylabel('x')
plt.title('Closed-loop behavior using CLF (c=2)')
plt.show
```

```python
# For Sontag's controller
def g(x, u):
    dgdt = -(x**3)*np.sqrt(1 + x**4)
    return dgdt

# Solve the ODE with the initial conditions 
x4 = odeint(g, x0_1, t)
x5 = odeint(g, x0_2, t)
x6 = odeint(g, x0_3, t)

# Plot x vs. t
plt.plot(t, x4, 'b', label = 'x(0) = 1')
plt.plot(t, x5, 'r', label = 'x(0) = 4')
plt.plot(t, x6, 'g', label = 'x(0) = 7')
plt.legend(loc='best')
plt.xlabel('t')
plt.ylabel('x')
plt.title('Closed-loop behavior using Sontag formula (c=2)')
plt.show
```
