```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
# Define the first order ODE 
def f(x, t):
    dxdt = -x**3
    return dxdt
```

```python
# Solve the ODE
x0 = 1
x = odeint(f, x0, t)
```

```python
# Lyapunov function 
V = 0.5*x**2
```

```python
# Plot V(x(t)) and x(t) vs. t
t = np.linspace(0, 100, 30)

plt.plot(t, x, 'b', label = 'x(t)')
plt.plot(t, V, 'r', label = 'V(x(t))')
plt.legend(loc='best')
plt.xlabel('t')
plt.ylabel('V(x(t)) and x(t)')
plt.savefig('Problem 6')
plt.show()
```
